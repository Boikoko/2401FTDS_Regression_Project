#### HANDLING OUTLIERS

```python
def find_outliers(df, threshold=3):
    """
    Detect outliers in numerical columns of a DataFrame using z-scores. Parameters: df (DataFrame): Input DataFrame containing numerical columns. threshold (float): Number of standard deviations to consider as outliers (default: 3). Returns: outliers (list of dicts): List of dictionaries containing outlier detection results for each column.

    Initialize an empty list to store outlier detection results. Iterate over each numerical column in the DataFrame. Calculate mean and standard deviation for the column. Calculate z-scores for each value in the column. Define outlier condition using z-scores. Extract potential outliers based on the defined condition. Store outlier detection results in a dictionary. Append the results to the list of outliers. Return the list of outlier detection results.
    """

    outliers = []
    for column in df.select_dtypes(include=['float64', 'int64']).columns:
        mean = df[column].mean()
        std_dev = df[column].std()
        z_scores = (df[column] - mean) / std_dev
        outlier_condition = (z_scores > threshold) | (z_scores < -threshold)
        potential_outliers = df[column][outlier_condition]
        outlier_count = len(potential_outliers)
        results = {
            'Column': column,
            'Count of Outliers': outlier_count,
            'Mean': mean,
            'Standard Deviation': std_dev,
            'Potential Outliers': potential_outliers.tolist()
        }
        outliers.append(results)
    return outliers
```
